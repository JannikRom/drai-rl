# Simple Environment 

```python
import json
import os
import glob
import matplotlib.pyplot as plt
import numpy as np
```

## Pendulum

```python


folder_path = "../plots/report_3_1/pendulum/rewards"  

json_files = sorted(glob.glob(os.path.join(folder_path, "*rewards.json")))

fig, ax = plt.subplots(figsize=(12, 6))

for filepath in json_files:
    with open(filepath, "r") as f:
        data = json.load(f)
    
    data = np.array(data)
    steps = data[:, 1]
    rewards = data[:, 2]
    label = os.path.splitext(os.path.basename(filepath))[0]
    
    ax.plot(steps, rewards, alpha=0.4, linewidth=0.8)
    
    window = max(1, len(rewards) // 30)
    smoothed = np.convolve(rewards, np.ones(window) / window, mode="valid")
    ax.plot(steps[window//2 : window//2 + len(smoothed)], smoothed, linewidth=2, label=label)

ax.set_xlabel("Steps")
ax.set_ylabel("Reward")
ax.set_title("Training Rewards")
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"{len(json_files)} Files: {[os.path.basename(f) for f in json_files]}")


print(f"{'Algorithmu':<30} | {'Final Reward (Avg 100 ± std)':<30}")
print("-" * 65)

for filepath in json_files:
    with open(filepath, "r") as f:
        data = json.load(f)
    
    data = np.array(data)
    rewards = data[:, 2]
    
    last_100 = rewards[-100:]
    
    avg = np.mean(last_100)
    std = np.std(last_100)
    
    label = os.path.splitext(os.path.basename(filepath))[0]
    
    print(f"{label:<30} | {avg:>10.2f} ± {std:<10.2f}")


```
